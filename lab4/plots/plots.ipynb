```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
timings_2000 = np.array([
    [1, 19.74],
    [2, 11.12],
    [4, 6.89],
    [8, 4.63],
    [16, 3.48],
])

timings_10000 = np.array([
    [1, 97.18],
    [2, 51.75],
    [4, 32.30],
    [8, 17.79],
    [16, 10.85],
])

timings_50000 = np.array([
    [1, 565.55],
    [2, 294.83],
    [4, 148.17],
    [8, 80.35],
    [16, 47.48],
])
```

```python
def plot(ax, timings, title):
    ax.set_title(title)
    ax.set_xlabel('Processors')
    ax.set_ylabel('Execution time (s)')
    ax.plot(timings[:, 0], timings[:, 1])
    ax.set_xticks(timings[:, 0])
    ax.set_yticks(timings[:, 1])
    ax.grid(True)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
plt.subplots_adjust(top=0.8)
fig.suptitle('Execution time')
plot(ax1, timings_2000, '2000 points')
plot(ax2, timings_10000, '10000 points')
plot(ax3, timings_50000, '50000 points')
fig.show()
```

```python
plt.title('Execution time')
plt.xlabel('Processors')
plt.ylabel('Execution time (s)')
for timings, points in [[timings_2000, 2000], [timings_10000, 10000], [timings_50000, 50000]]:
    plt.plot(timings[:, 0], timings[:, 1], label='{} points'.format(points))
plt.grid(True)
plt.legend()
```
