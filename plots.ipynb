```python
import matplotlib.pyplot as plt
```

```python
cpu_3x3 = np.array([
    [1000**2, 0.069],
    [2000**2, 0.314],
    [2500**2, 0.49],
    [2671*4006, 0.84],
])
gpu_3x3 = np.array([
    [1000**2, 0.163],
    [2000**2, 0.183],
    [2500**2, 0.226],
    [2671*4006, 0.276],
])

cpu_9x9 = np.array([
    [1000**2, 0.602],
    [2000**2, 1.954],
    [2500**2, 3.777],
    [2671*4006, 6.448],
])
gpu_9x9 = np.array([
    [1000**2, 0.173],
    [2000**2, 0.195],
    [2500**2, 0.245],
    [2671*4006, 0.416],
])

cpu_15x15 = np.array([
    [1000**2, 1.651],
    [2000**2, 6.642],
    [2500**2, 10.351],
    [2671*4006, 16.755],
])
gpu_15x15 = np.array([
    [1000**2, 0.237],
    [2000**2, 0.309],
    [2500**2, 0.306],
    [2671*4006, 0.351],
])

cpu_21x21 = np.array([
    [1000**2, 3.186],
    [2000**2, 12.729],
    [2500**2, 19.849],
    [2671*4006, 34.132],
])
gpu_21x21 = np.array([
    [1000**2, 0.252],
    [2000**2, 0.354],
    [2500**2, 0.578],
    [2671*4006, 0.541],
])
```

```python
def plot(ax, gpu, cpu, title):
    ax.set_title(title)
    ax.set_xlabel('Pixels')
    ax.set_ylabel('Execution time (s)')
    ax.plot(cpu[:, 0], cpu[:, 1], label='cpu')
    ax.plot(gpu[:, 0], gpu[:, 1], label='gpu')
    ax.axis([40, max(cpu[:, 0]) + 1000000, 0, max(cpu[:, 1]) + 0.5])
    ax.grid(True)
    ax.legend()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(15,3))
plt.subplots_adjust(top=0.8)
fig.suptitle('CPU vs GPU')
plot(ax1, gpu_3x3, cpu_3x3, 'Blur with 3x3 filter matrix')
plot(ax2, gpu_9x9, cpu_9x9, 'Blur with 9x9 filter matrix')
plot(ax3, gpu_15x15, cpu_15x15, 'Blur with 15x15 filter matrix')
plot(ax4, gpu_21x21, cpu_21x21, 'Blur with 21x21 filter matrix')
fig.show()
```

```python
plt.title('Execution time comparison with different filter matrix sizes')
plt.xlabel('Filter matrix size')
plt.ylabel('Execution time (s)')
plt.plot([3, 9, 15, 21], [gpu_3x3[-1, 1], gpu_9x9[-1, 1], gpu_15x15[-1, 1], gpu_21x21[-1, 1]])
plt.grid(True) 
plt.show()
```
